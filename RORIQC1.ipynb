```python
file_path = '/Users/sameeabutt/Desktop/Rori UPenn/Code/RORI_survey_responses_2023-12-01-2024-01-07.csv'

method = 'zeroshot'

candidate_labels = [f"Provide better explanations", "offer more examples", "add more subjects", "show more personality", "add more program features"]

def zero_shot_classification(prompt, labels):
    classification_prompt = f"Please classify the following text into one of the following categories: {', '.join(labels)}.\n\nText: \"{prompt}\"\nCategory:"
    try:
        response = openai.Completion.create(
        engine= "gpt-4-turbo-2024-04-09",
        prompt=classification_prompt,
        max_tokens=200,
        n=1,
        stop=["\n"],
        temperature=0
    )
        result = response.choices[0].text.strip()
        return result
    except Exception as e:
        print("An error occurred during the API call:", str(e))
        return None
```
